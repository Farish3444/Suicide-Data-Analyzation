```python
# All Important Libraries to Perform Exploratory Data Analysis
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import cufflinks as cf
import plotly as py
import plotly.express as px
import plotly.graph_objects as go
py.offline.init_notebook_mode(connected=True)
cf.go_offline()
import os
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
le=LabelEncoder()
ohe=OneHotEncoder()
```

```python
data = pd.read_csv("D:\KAGGLE DATASET\Suicide Rates Overview 1985 to 2016\master.csv")
```

```python
data.rename(columns={'suicides/100k pop':'suicides/100k',' gdp_for_year ($) ':'gdp_for_year','gdp_per_capita ($)':'gdp_per_capita'},inplace=True)
```

```python
data.head()
```

```python
# Data Visualization on Generation
# Code for Bar Graph
data['generation'].iplot(kind='hist')
```

```python
# Code for Pie Graph
generationpie = data.groupby('generation').size().reset_index(name='numbers').sort_values('numbers',ascending=True)
px.pie(generationpie,names='generation',values='numbers')
```

```python
# Countries with Least Number of Suicide Rates
cou = data.groupby('country').size().reset_index(name='Total').sort_values('Total',ascending=True).head(11)
px.bar(cou,x='country',y='Total')
```

```python
plt.figure(figsize=(16,7))
bar_show = sns.barplot(x = 'sex', y = 'suicides_no', hue = 'age',data = data)
```

```python
# Data from 20th Century.
age=data.groupby('age').size().reset_index(name='nos').sort_values('nos',ascending=False)
px.pie(age,names='age',values='nos')
```

```python
latest=data[data['year']==2015]
latest_age = latest.pivot_table(index=('age'),columns='country',values='suicides/100k').reset_index()
latest_age['age'] = pd.Categorical(latest_age['age'], ['5-14 years','15-24 years','25-34 years','35-54 years','55-74 years','75+ years'])
latest_age = latest_age.sort_values('age')

fig, axes = plt.subplots(ncols=2, sharey=True,sharex=True,figsize=(15,10))


axes[0].barh('age','Turkey', align='center', data=latest_age, zorder=10)
axes[0].set(title='Turkey',xlabel='Suicides/100k')
axes[1].barh('age', 'Others', align='center',data=latest_age, color='gray', zorder=10)
axes[1].set(title='Others',xlabel='Suicides/100k')

# axes[1].invert_xaxis()
# axes[0].set(yticks=y, yticklabels=latest_age['age'].unique)
axes[0].yaxis.tick_right()

for ax in axes.flat:
    ax.margins(0.03)
    ax.grid(True)
    


fig.tight_layout()
fig.subplots_adjust(wspace=0.15)
plt.show()
```
